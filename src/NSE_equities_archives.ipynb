```python
import csv
import requests
import pandas as pd



```

```python
CSV_URL = 'https://www1.nseindia.com/archives/equities/mto/MTO_{}.DAT'
```

```python
from datetime import date
import datetime

today = date.today()
# t
```

```python
# base = datetime.datetime.today()
numdays = 365
today = date.today()
date_list = [today - datetime.timedelta(days=x) for x in range(numdays)]
```

```python
dates = []
for aDay in date_list:
    today = aDay.strftime("%d%m%Y")
    filename = str(today)
    dates.append(filename)
#     print(CSV_URL.format(filename))
    
```

```python
for aDay in dates:
    response = requests.get(CSV_URL.format(aDay)) 
    if response.status_code == 200:
        with open('/home/jupyter/data/nse' +aDay+'.csv', 'w') as f:
            writer = csv.writer(f)
            iterator = response.iter_lines()
            print(next(iterator))
            print(next(iterator))
            print(next(iterator))
            for line in response.iter_lines():
                writer.writerow(line.decode('utf-8').split(','))
    else:
        print(CSV_URL)
```
